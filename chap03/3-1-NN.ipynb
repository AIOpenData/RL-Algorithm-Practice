### 神经元的激活函数：阶跃函数

```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
```

```python
def unit(x):
    # sgn
    if x>=0:
        return 1
    else:
        return 0
```

```python
line_x = np.linspace(-5, 5, 100)
line_y = []
for i in line_x:
    line_y.append(unit(i))
plt.xlabel("z")
plt.ylabel("u(z)")
plt.xlim(-5,5)     
plt.ylim(-0.1,1.1) 
plt.plot(line_x, line_y) 
plt.savefig("fig_chap8_unit.jpg", dpi=600)
plt.show()
```

### 神经元的激活函数：Sigmoid函数

```python
def sigmoid(x):
    # sigmoid function
    return 1/(1 + np.exp(-x))
```

```python
line_x = np.linspace(-5, 5, 100)
line_y = sigmoid(line_x)

plt.xlabel("z")
plt.ylabel("sigmoid(z)")
plt.xlim(-5,5)     
plt.ylim(0,1) 
plt.plot(line_x, line_y, label='Sigmoid') #, c='blue', label='Sigmoid'
plt.scatter([0, ], [0.5, ], s=20,color='black') #color='r'
# plt.plot([0,0,],[0,0.5],'k--',color='black',linewidth=1)   参数重复定义，fmt='k--'的含义为黑色虚线，所以后面color='black'是冗余的
# plt.plot([-5,0,],[0.5,0.5],'k--',color='black',linewidth=1)
plt.plot([0,0,],[0,0.5],'k--',linewidth=1)
plt.plot([-5,0,],[0.5,0.5],'k--',linewidth=1)
plt.annotate(r'(0,0.5)', xy=(0.2, 0.5), xycoords='data',fontsize=12)
plt.legend()
plt.savefig("fig_chap8_sigmoid.jpg", dpi=600)
plt.show()
```

### 感知机（Perceptron）的激活函数：符号函数

```python
def sgn(x):
    # sgn
    if x>=0:
        return 1
    else:
        return -1
```

```python
line_x = np.linspace(-5, 5, 100)
line_y = []
for i in line_x:
    line_y.append(sgn(i))
plt.xlabel("z")
plt.ylabel("sgn(z)")
plt.xlim(-5,5)     
plt.ylim(-1.1,1.1) 
plt.plot(line_x, line_y, label='Sgn')
# plt.plot([-5,5,],[0,0],'k--',color='black',linewidth=1)    color='black'冗余
plt.plot([-5,5,],[0,0],'k--',linewidth=1)
plt.legend()
plt.savefig("fig_chap8_sgn.jpg", dpi=600)
plt.show()
```
